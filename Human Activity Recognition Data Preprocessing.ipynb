# Mapping the Activity names to the data

```python
activity_labels =pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\activity_labels.txt', sep = "\s+", names = ['id','label_names'])
activity_labels
```

```python
# Creating a Dictionary to assign the values in the y-train data to the activity name
activitymap = {str(row.id):row.label_names for i, row in activity_labels.iterrows()}
activitymap
```

# Importing Train dataset

```python
trainx = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\train\X_train.txt',sep = '\s+',   header = None
                    , mangle_dupe_cols = True, encoding = 'latin-1')


trainx.columns = features.feature_name.tolist()
```

# Importing and finessing test data


```python
test = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\test\X_test.txt',sep = '\s+',   header = None
                    , mangle_dupe_cols = True, encoding = 'latin-1')
act_y = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\test\y_test.txt',sep = '\s+',   names = ['Activity'])
test.columns = features.feature_name.tolist()
act_y
```

```python
act_x = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\train\y_train.txt',sep = '\s+',   names = ['Activity'])
act_x
```

# Merging datasets with features

```python
test = pd.concat([act_y, test], axis = 1)
train = pd.concat([act_x, trainx], axis = 1)
test
```

```python
data  = pd.concat([train, test], axis = 0)
```

```python
data.shape
```

# Mapping Activity Labels to each activity

```python
data['Activity'] = data['Activity'].apply(lambda x: activitymap[str(x)])
data.head(5)
```

# Creating New dataframe containing only mean and std data

```python
filth = data.filter(regex = 'mean|std')
filth = pd.concat([data['Activity'],filth], axis = 1)
pd.set_option('display.max_columns', None )
filth
```

# Modifying Descriptive Variable Name

```python
for i in [('tBody' , 'TimeBody'), ('mean()', 'Mean'),('tGravity','TimeGravity'),('std()', 'Standard Deviation'),('fBody' ,'FrequencyBody'), ('Acc', 'Acceleration')
                                 , ('Gyro', 'Gyroscopic'), ('Mag' ,'Magnitude'), ('Freq()' , 'Frequency'),('Bodybody' , 'Body')]:
    filth.columns = filth.columns.str.replace(*i, regex = False)
```

```python
filth
```

```python
subject_train = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\train\subject_train.txt', sep = "\s+", names = ['Subject ID'])
subject_test = pd.read_csv(r'C:\Users\PAVILLION\Desktop\UCI HAR Dataset\test\subject_test.txt', sep = "\s+", names = ['Subject ID'])
```

```python
sub = pd.concat([subject_train, subject_test], axis = 0)
sub
```

```python
semis = pd.concat([sub, filth], axis =1)
semis
```

```python
final = semis.groupby(['Subject ID','Activity']).mean()
final
```
